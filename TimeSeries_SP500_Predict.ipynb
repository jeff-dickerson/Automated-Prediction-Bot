```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) #converted to integer
```

```python
sp500
```

Dataframe shows Target column indicating 1 if Tomorrows price is greater than todays price


```python
sp500 = sp500.loc["1990-01-01":].copy()
```

Figured going back to far would not be as useful for future predictions due to shifts in conditions and variables so I removed all data before 1990

# Training An Initial Machine Learning Model

Start by using Random Forest Classifier which is best for *non-linear data relationships*.
It works by training a bunch of individual decision trees with randomized parameters and than averaging the results from those decision trees

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100]

predictors = ["Close", "Volume", "Open" , "High", "Low"]
model.fit(train[predictors], train["Target"])
```

Because this is time-series data I split it up into test set to avoid using future data to predict the past which is ridiculous in the real world, causes data leakage (Leakage means leaking info into the model)

```python
from sklearn.metrics import precision_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(train[predictors], train["Target"])

test[predictors] = test[predictors].values.reshape(-1, len(predictors))
preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```
